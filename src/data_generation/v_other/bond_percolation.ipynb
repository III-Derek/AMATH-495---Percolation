```julia
mutable struct Cluster
    size::Int
    parent::Union{Cluster, Nothing}
    Cluster(size,) = new(size, nothing)
end

struct Node
    cluster::Cluster
end

struct Bond
    node1::Node
    node2::Node
end


function remove_bond()

end
function create_square_lattice(L,W)
    nodes = Array{Node}(undef, (L, W))
    for i in 1:L
        for j in 1:W
            nodes[i,j] = Node(Cluster(1))
        end
    end
    bond_list::Vector{Bond} = []
    for i in 1:L-1
        for j in 1:W-1
            bond = Bond(nodes[i,j], nodes[i,j+1])
            push!(bond_list, bond)
            bond = Bond(nodes[i,j+1], nodes[i,j])
            push!(bond_list, bond)
        end
    end
    for i in 1:L-1
        bond = Bond(nodes[i,W], nodes[i+1,W])
        push!(bond_list, bond)
    end
    for i in 1:W-1
        bond = Bond(nodes[L,i], nodes[L,i+1])
        push!(bond_list, bond)
    end
    return bond_list
end

max_size = 1
bond_list = create_square_lattice(120,120)
largest_cluster::Vector{Int} = zeros(length(bond_list))
for iteration in 1:length(bond_list)
    i = rand(1:length(bond_list))
    cluster1 = bond_list[i].node1.cluster
    while !isnothing(cluster1.parent)
        cluster1 = cluster1.parent
    end
    cluster2 = bond_list[i].node2.cluster
    while !isnothing(cluster2.parent)
        cluster2 = cluster2.parent
    end
    if cluster1 != cluster2
        new_size = cluster1.size + cluster2.size
        new_cluster = Cluster(new_size)
        cluster1.parent = new_cluster
        cluster2.parent = new_cluster
        if new_size > max_size
            max_size = new_size
        end
    end
    largest_cluster[iteration] = max_size
    deleteat!(bond_list, i)
end
```

```julia
print("[")
for i in 1:length(largest_cluster)
    print(largest_cluster[i],",")
end
print("]")
```
