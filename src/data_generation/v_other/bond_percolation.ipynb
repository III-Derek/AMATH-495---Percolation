```julia
mutable struct Cluster
    size::Int
    parent::Union{Cluster, Nothing}
    Cluster(size,) = new(size, nothing)
end

struct Node
    cluster::Cluster
end

struct Bond
    node1::Node
    node2::Node
end


function remove_bond()

end
function create_square_lattice(L,W)
    nodes = Array{Node}(undef, (L, W))
    for i in 1:L
        for j in 1:W
            nodes[i,j] = Node(Cluster(1))
        end
    end
    bond_list::Vector{Bond} = []
    for i in 1:L-1
        for j in 1:W-1
            bond = Bond(nodes[i,j], nodes[i,j+1])
            push!(bond_list, bond)
            bond = Bond(nodes[i,j+1], nodes[i,j])
            push!(bond_list, bond)
        end
    end
    for i in 1:L-1
        bond = Bond(nodes[i,W], nodes[i+1,W])
        push!(bond_list, bond)
    end
    for i in 1:W-1
        bond = Bond(nodes[L,i], nodes[L,i+1])
        push!(bond_list, bond)
    end
    return bond_list
end

max_size = 1
bond_list = create_square_lattice(120,120)
largest_cluster::Vector{Int} = zeros(length(bond_list))
for iteration in 1:length(bond_list)
    i = rand(1:length(bond_list))
    cluster1 = bond_list[i].node1.cluster
    while !isnothing(cluster1.parent)
        cluster1 = cluster1.parent
    end
    cluster2 = bond_list[i].node2.cluster
    while !isnothing(cluster2.parent)
        cluster2 = cluster2.parent
    end
    if cluster1 != cluster2
        new_size = cluster1.size + cluster2.size
        new_cluster = Cluster(new_size)
        cluster1.parent = new_cluster
        cluster2.parent = new_cluster
        if new_size > max_size
            max_size = new_size
        end
    end
    largest_cluster[iteration] = max_size
    deleteat!(bond_list, i)
end
```

```julia
using FileIO
using Images
img = FileIO.load("C:\\Users\\jonat\\OneDrive\\Documents\\programming\\AnacondaProjects\\PHYS495\\final\\data\\perco_image\\0.976_60863.png");
imgarr = channelview(img)
n_columns = size(imgarr, 1)
n_rows = size(imgarr,2)

largest_label = 0;
label = zeros(n_columns, n_rows)
labels = collect(1:n_columns*n_rows) # Array containing integers from 0 to the size of the image. 

function combine(x::Int, y::Int)
    labels[find(x)] = find(y);
end

function find(x::Int)
    y::Int = x;

    while (labels[y] != y)
        y = labels[y];
    end

    while (labels[x] != x) 
        z::Int = labels[x];
        labels[x] = y;
        x = z;
    end

    return y;
end

for x in 1:n_columns
    for y in 1:n_rows
        if imgarr[x, y] == 1
            left = x > 1 ? floor(Int,label[x-1, y])  : 0
            above = y > 1 ? floor(Int,label[x, y-1]) : 0
            if (left == 0) && (above == 0) # Neither a label above nor to the left.
                largest_label = largest_label + 1; # Make a new, as-yet-unused cluster label.
                label[x, y] = largest_label;
            elseif (left != 0) && (above == 0) #  One neighbor, to the left.
                label[x, y] = find(left);
            elseif (left == 0) && (above != 0)# One neighbor, above. 
                label[x, y] = find(above);
            else # Neighbors BOTH to the left and above. 
                combine(left,above); #Link the left and above clusters. 
                label[x, y] = find(left);
            end
        end
    end
end


```
